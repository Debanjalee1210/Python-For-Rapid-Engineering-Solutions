```python
#Creating arrays
import numpy as np
#easy way to create arrays is specify the size with a convenience of numpy
array_0 = np.zeros(4,float) # need to specify the type of value
array_1 = np.ones(4,complex) #any type works
array_2 = np.full([2,3],23) #can specify an initial value for  all entries
array_3 = np.full_like(array_2,2) #make new array same as other but new values 

print(array_0)
print(array_1)
print(array_2)
print(array_3)
```

```python
#provide a list as an argument
mk_array = np.array([11,12.9,13],int)
print("make array",mk_array)

mk_array = np.array([11,12.9,13],float)
print("make array",mk_array)

#provide a lsit variable
my_list = [22,23,24]
my_list_array = np.array(my_list,int)
print("list array",my_list_array)

#careful of empty and empty_like - they will be filled with whatever
#happened to be in those memory locations.This will look like randome garbage value!!
array_4 = np.empty(8,float)
print("Empty array :",array_4)
```

```python
#Modifying array entries
array_4  = np.zeros(4,int)
print(array_4)
# add,multiply  entry by entry
array_4 += 2
print(array_4)

array_4 *= 3
print(array_4)
```

```python
#copying arrays
array_5 = np.zeros(4,int)
array_6 = np.ones(4,int)
print(array_6)
array_6 = array_5 #points to same memory
print(array_6)

array_7 = np.copy(array_5)
array_7[0] = 23
print("\nno change",array_5)
print(array_7)
```

```python
#Numerical integration
#example quad integration
import numpy as np
from scipy.integrate import quad 
import matplotlib.pyplot as plt

POINTS = 10  # number of values of 'a' to use

def my_func(x,a):
    y = np.exp(a*x)
    return y

#initialize arrays to hold the values of a and the integral results
a_vals = np.zeros(POINTS,int)
integrals = np.zeros(POINTS,float)

print("a value error")
for a_val in range(POINTS):
    a_vals[a_val] = a_val 
    
    integrals[a_val],err = quad(my_func,0,1,args =(a_val))
    
    print(" ",a_val," {:.2e}".format(err))
    
    plt.plot(a_vals,integrals)
    plt.xlabel("value of 'a'")
    plt.ylabel("e^(ax)")
    plt.title('example 1')
    plt.show()
```
